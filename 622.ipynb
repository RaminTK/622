```python
from ipyleaflet import Map, Marker, basemaps
```

```python
# Define the base Google Maps layer
google_map = basemaps.Gaode.Satellite

# Create the map centered at an initial location
initial_location = (df['LATITUDE'].mean(), df['LONGITUDE'].mean())  # Replace 'lat' and 'long' with your column names
m = Map(basemap=google_map, center=initial_location, zoom=10)

# Add markers for each location in your DataFrame
for idx, row in df.iterrows():
    marker = Marker(location=(row['LATITUDE'], row['LONGITUDE']))  # Replace 'lat' and 'long' with your column names
    m.add_layer(marker)

# Display the map
print(m)
```
